```python
import pandas as pd
import seaborn as sns; sns.set(color_codes=True)
import numpy as np
```

```python
datafile = 'data/GDP.csv'
GDP_df = pd.read_csv(datafile)
GDP_df = GDP_df.dropna()
GDP_df.head()
```

```python
GDP_df['GDP'] = GDP_df['GDP'].astype('int64')
GDP_df['Time'] = GDP_df['Time'].astype('int64')
```

```python
GDP_df.head()
```

```python
datafile = 'data/econ_freedom.csv'
econ_df = pd.read_csv(datafile)
#GDP_df = GDP_df.dropna()
econ_df.head()
```

```python
merged = pd.merge(GDP_df, econ_df, left_on=['Time', 'country_code'], right_on=['year', 'ISO_Code'], how='outer')
```

```python
merged = merged.dropna()
merged
```

```python
data = merged.filter(['GDP','index'], axis=1)
```

```python
ax1 = data.plot.scatter(x='index',y='GDP',c='DarkBlue')
```

```python
lower = data[data.GDP < np.percentile(data.GDP,5)]
```

```python
ax1 = lower.plot.scatter(x='index',y='GDP',c='DarkBlue')
```
